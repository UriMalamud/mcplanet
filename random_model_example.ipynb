# Basic example:  

Create everything from scratch.   You specify mass, moment, radius, total rock/water/env, and amount of rock/env in the extreme shells. You need to rerun this cell to generate a new model (and rerun subsequent cell if you want to see the new results).

```python
model = create_mcinterior(mass, moment_ratio, radius, pct_rock, pct_env, num_shells, rock_0, env_0)
print("Generated random model")
print("Total mass is %s:"%model.get_mass())
print("Total moment is %s:"%model.get_moment())
print("Total mass/moment ratio %s"%model.get_mass_moment_ratio())
```
